```python
#taken from https://thinkdiff.net/how-to-run-javascript-in-python-web-scraping-web-testing-16bd04894360
import time
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
import os
import random


chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--window-size=1024x2400")

# download Chrome Webdriver  
# https://sites.google.com/a/chromium.org/chromedriver/download
# put driver executable file in the script directory
chrome_driver = os.path.join(os.getcwd(),  "headers/chromedriver")

driver = webdriver.Chrome(options=chrome_options, executable_path=chrome_driver)

driver.get("https://www.resultados.eleccionesgenerales2021.pe/EG2021/Actas/Numero/055407")

time.sleep(2+random.uniform(0, 1))

driver.find_element_by_xpath('//*[@id="pdf"]/div[3]/div[1]/ul/li[2]/a').click()

time.sleep(2+random.uniform(0, 1))



html=driver.page_source


# screenshot capture
driver.get_screenshot_as_file("python-scrap.png")
driver.close()
```
