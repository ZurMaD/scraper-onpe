```python
import pandas as pd
from urllib.request import Request, urlopen
import json
```

```python
URL='https://api.resultados.eleccionesgenerales2021.pe/mesas/detalle/044460'


req = Request(URL,     
              headers={'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36'})
webpage_json = urlopen(req).read().decode("utf-8")

print(type(webpage_json))

#print(webpage_json)

my_json=json.loads(webpage_json)
```

```python
df_acta_congresal=pd.DataFrame()


for i in range(20):
    my_row_from_json=my_json['procesos']['generalCon']['votos'][0]
    df_acta_congresal=df_acta_congresal.append(my_row_from_json, 
                         ignore_index=True)


```
